```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
n = 20
x_ = np.linspace(-5,5,n)
x_corr = x_ + np.random.randn(n)*0.0001
X = np.vstack((x_, x_corr)).T
rand_w = np.array([0.5, 0.3])
y = X *rand_w  + np.random.randn(n, 1)*0.2
```

```python
# Create matrix and vectors
n = 20
x_ = np.linspace(-5,5,n)
X = pd.DataFrame({'x': x_})
y = np.sin(x_) + np.random.randn(n)*0.3
```

```python
# construct the polynomial basis
X["x^2"] = x_**2
X["x^3"] = x_**3
X["x^4"] = x_**4
X["x^5"] = x_**5
```

```python
# view the original data
plt.scatter(x_, y)
```

```python
# linear regression on the original data
ncols = 1
regr = LinearRegression()
regr.fit(X.iloc[:,:ncols], y)

y_hat = regr.predict(X.iloc[:,:ncols])
plt.scatter(x_, y)
plt.plot(x_, y_hat)
```

```python
# linear regression on the quadratic base
ncols = 2
regr = LinearRegression()
regr.fit(X.iloc[:,:ncols], y)

y_hat = regr.predict(X.iloc[:,:ncols])
plt.scatter(x_, y)
plt.plot(x_, y_hat)
```

```python
# linear regression on the cubic base
ncols = 3
regr = LinearRegression()
regr.fit(X.iloc[:,:ncols], y)

y_hat = regr.predict(X.iloc[:,:ncols])
plt.scatter(x_, y)
plt.plot(x_, y_hat)
```

```python
# linear regression on the quartic base
ncols = 4
regr = LinearRegression()
regr.fit(X.iloc[:,:ncols], y)

y_hat = regr.predict(X.iloc[:,:ncols])
plt.scatter(x_, y)
plt.plot(x_, y_hat)
```

```python
# linear regression on the quintic base
ncols = 5
regr = LinearRegression()
regr.fit(X.iloc[:,:ncols], y)

y_hat = regr.predict(X.iloc[:,:ncols])
plt.scatter(x_, y)
plt.plot(x_, y_hat)
```

### Be careful when extrapolating outside of your training set

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# PolynomialFeatures (prepreprocessing)
poly = PolynomialFeatures(degree=5, include_bias = False)
X_poly = poly.fit_transform(np.linspace(-20,20,200).reshape(-1,1))
```

```python
# linear regression on the quintic base
y_hat = regr.predict(X_poly)
plt.scatter(x_, y)
plt.plot(X_poly[:,0], y_hat)
plt.yscale("symlog")
```

### Also be careful when using high order polynomials

```python
poly = PolynomialFeatures(degree=17, include_bias = False)
X_poly = poly.fit_transform(x_.reshape(-1,1))
# linear regression on the quintic base
regr.fit(X_poly, y)

X_poly = poly.fit_transform(np.linspace(-5,5,200).reshape(-1,1))
y_hat = regr.predict(X_poly)
plt.scatter(x_, y)
plt.plot(X_poly[:,0], y_hat)
plt.yscale("symlog")
```
